Dataset Description:
This report analyzes state probability transaction data between different Indian states. The dataset consists of 315 rows and 7 columns, including home state, sale state, transaction count, and (month, year).

The columns in our dataset are homestate code, salestate code, month, year.

# Using PairGrid() method forward with homestatecode,salestatecode,txn_count,Total codes,month

```python
import matplotlib.pyplot as plt
import seaborn as sns
home_state_summary = df.groupby('homestatename')['txn_count'].agg(['sum', 'mean']).rename(columns={'sum': 'Total Transactions', 'mean': 'Average Transactions'})
sale_state_summary = df.groupby('salestatename')['txn_count'].agg(['sum', 'mean']).rename(columns={'sum': 'Total Transactions', 'mean': 'Average Transactions'})
df
```

# Transaction Analysis: Total and Average Transactions by State

```python
time_summary = df.groupby(['year', 'month'])['txn_count'].sum().reset_index()
time_summary['Date'] = pd.to_datetime(time_summary[['year', 'month']].assign(day=1))
df
```

# Temporal Analysis: Transactions by Month and Year

```python
state_pair_summary = df.groupby(['homestatename', 'salestatename'])['txn_count'].sum().reset_index()
state_pair_summary = state_pair_summary.sort_values(by='txn_count', ascending=False).head(10)
df
```

# State Pair Analysis

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=time_summary, x='Date', y='txn_count')
plt.title('Monthly Transaction Trends')
plt.xlabel('Date')
plt.ylabel('Transaction Count')
plt.show()
df
```

# Visualization: Monthly Transaction Trends

```python
#19
plt.figure(figsize=(12, 6))
sns.barplot(data=state_pair_summary, x='txn_count', y='homestatename', orient='h')
plt.title('Top 10 State Pairs by Transactions')
plt.xlabel('Transaction Count')
plt.ylabel('Home State Name')
plt.show()
```

# Visualization: Top State Pairs by Transactions

```python
#1
plt.figure(figsize=(12, 6))
sns.barplot(data=sale_state_summary.reset_index(), x='salestatename', y='Total Transactions')
plt.title('Total Transactions by Sale State')
plt.xlabel('Sale State')
plt.ylabel('Total Transactions')
plt.xticks(rotation=90)
plt.show()
```

#  Additional Visualizations  Total Transactions per Sale State

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=home_state_summary.reset_index(), x='homestatename', y='Average Transactions')
plt.title('Average Transactions by Home State')
plt.xlabel('Home State')
plt.ylabel('Average Transactions')
plt.xticks(rotation=90)
plt.show()
```

# Average Transactions per Home State

```python
#4
state_matrix = df.pivot_table(index='homestatename', columns='salestatename', values='txn_count', aggfunc='sum', fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(state_matrix, cmap='coolwarm', annot=True, fmt='d')
plt.title('Heatmap of Transactions Between States')
plt.xlabel('Sale State')
plt.ylabel('Home State')
plt.show()
```
